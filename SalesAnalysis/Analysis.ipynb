# Sales Analysis

```python
import pandas as pd
import os
```

## Task #1: Merging 12 months of sales data into a single CSV file

```python
files = [file for file in os.listdir('./Sales_Data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, df])
    
all_months_data.to_csv("all_data.csv", index=False)
    
```

### Read in updated dataframe

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

### Task #2: Cleaning the data

#### Drop rows of NaN

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

all_data = all_data.dropna(how='all')
all_data.head()
```

#### Find 'Or' and Delete it

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

#### Convert columns to the correct type (all are currently as type string)

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) # Convert to int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) # Convert to Float
```

## Augment data with additional columns

### Task #3: Add Month Column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

### Task #4: Add a Sales Column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

### Task #5: Add a City Column

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' (' + get_state(x) + ')')
all_data.head()
```

## Question 1: What was the best month for sales? How much was earned that month? Worst month?

```python
results = all_data.groupby('Month').sum()
results
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months, results['Sales'])
plt.ticklabel_format(style='plain') # to prevent scientific notation
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month')
plt.show()

# December was the best month for sales ($4613443.34), and January was the worst month for sales ($1822256.73).
```

## Question 2: What city had the highest number of sales? Lowest?

```python
results = all_data.groupby('City').sum()
results
```

```python
import matplotlib.pyplot as plt

cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation='vertical', size=8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City')
plt.show()

# San Francisco (CA) had the highest number of sales ($8262203.91), and Portland (ME) had the lowest number
# of sales ($449758.27).
```

## Question 3: What time should we display advertisements in order to maximize likelihood of customer's buying product?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hour of Day')
plt.ylabel('Number of Purchases')
plt.grid()
plt.show()

# The number of orders peak at around 11am and 7pm, so these times are optimal for advertising.
```

## Question 4: What products are most often sold together?

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ', '.join(x))

df = df[['Order ID', 'Grouped']].drop_duplicates()

df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(10):
    print(key, value)
    

# The pair of products that are most often sold together are an iPhone and a Lightning Charging Cable.
```

## Question 5: What product sold the most? Why?

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

products = [product for product, df in product_group]

plt.bar(products, quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product Name')
plt.xticks(products, rotation='vertical', size=8)
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']

fit, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

plt.show()

# AAA Batteries (4-pack) sold the most. Less expensive products, such as AAA batteries, generally sold more, while more expensive products
# generally sold less. 
```
